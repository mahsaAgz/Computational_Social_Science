```python
# prompt: Load in_class_lab1_Trump_Tweets.csv

import pandas as pd

# Load the dataset
df = pd.read_csv("Trump_Tweets.csv", encoding="utf-8")

# Print the first five rows
df.head()
```

```python
# prompt: Transform the timestamp information in "created_at" variable into the real date that humans can recognize by using the function utcfromtimestamp in datetime package.

import datetime

df["created_at"] = df["created_at"].apply(lambda x: datetime.datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
df.head()
```

```python
import pandas as pd
from gensim import corpora, models
import gensim
import nltk
from nltk.corpus import stopwords
import re

# Ensure you have the necessary NLTK data
nltk.download('stopwords')

# Load the dataframe
# df = ... # Load your dataframe here

# Preprocessing the text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove URLs, handles, and the hashtag symbol
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@\w+|\#','', text)
    # Keep only alphabets
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # Remove short words (less than 3 characters)
    text = re.sub(r'\b\w{1,2}\b', '', text)
    # Remove extra spaces
    text = re.sub(r'\s\s+', ' ', text)
    return text

# Apply preprocessing to each tweet
df['clean_text'] = df['text'].map(preprocess_text)

# Tokenize the text and remove stop words
stop_words = set(stopwords.words('english'))
df['tokens'] = df['clean_text'].apply(lambda x: [word for word in x.split() if word not in stop_words])

# Create a dictionary and corpus needed for topic modeling
dictionary = corpora.Dictionary(df['tokens'])
corpus = [dictionary.doc2bow(text) for text in df['tokens']]

# Apply LDA
lda_model = gensim.models.ldamodel.LdaModel(corpus, num_topics=15, id2word=dictionary, passes=30,  random_state=100)
lda_model.print_topics()

```

```python
from textblob import TextBlob

# Function to get the polarity of each tweet
def get_tweet_sentiment(tweet):
    # Create a TextBlob object
    analysis = TextBlob(tweet)
    # return sentiment
    return analysis.sentiment.polarity

# Apply function to get sentiment of each tweet
df['sentiment'] = df['clean_text'].apply(get_tweet_sentiment)
```

```python
# Assigning topics to each tweet
def assign_topic(row):
    bow = dictionary.doc2bow(row)
    topic_scores = lda_model.get_document_topics(bow)
    # Select the topic with highest contribution
    main_topic = sorted(topic_scores, key=lambda x: x[1], reverse=True)[0][0]
    return main_topic

# Apply the function to the dataframe
df['main_topic'] = df['tokens'].apply(assign_topic)

# Group by main topic and calculate average likes and retweets
topic_engagement = df.groupby('main_topic').agg({
    'favorite_count': 'mean',
    'retweet_count': 'mean'
}).reset_index()

print("Average Likes and Retweets by Topic")



topic_names = {
0: "Presidential Greetings and Ceremonies",
1: "International Trade and Diplomacy",
2: "Patriotism and National Pride",
3: "Public Appearances and Messages",
4: "Evening Events and Rallies",
5: "Economic Policies and Health Care",
6: "Media Critique and International Concerns",
7: "International Relations and Security",
8: "Economic Achievements and Records",
9: "Tax Policy and Healthcare Reform",
10: "Social Issues and Governance",
11: "Community Engagement and Support",
12: "Media Criticism",
13: "Political Strategy and Regulation",
14: "International Politics and Allegations"
}


topic_engagement['topic_name'] = topic_engagement['main_topic'].map(topic_names)
print(topic_engagement)
```
