```python
# Assuming you want to create an empty DataFrame with specified column names
columns = ['SNS_name', 'pew_estimate', 'our_estimate']
df3 = pd.DataFrame(columns=columns)

social_media_columns = ['Do you ever use WhatsApp?', 'Do you ever use  Pinterest?', 'Do you ever use TikTok?',
                        'Do you ever use Instagram?', 'Do you ever use Snapchat?', 'Do you ever use Facebook?',
                        'Do you ever use LinkedIn?', 'Do you ever use Youtube?', 'Do you ever use Twitter (X)?',
                        'Do you ever use Reddit?','Do you ever use BeReal?']

# Calculate Pew estimates for each social media platform
pew_estimates = {}
for column in social_media_columns:
    platform_name = column.split('Do you ever use ')[1].split('?')[0]  # Extracting platform name
    yes_count = df[column].str.lower().eq('yes').sum()  # Counting 'Yes' responses
    total_count = df[column].notnull().sum()  # Counting total non-null responses
    pew_estimate = yes_count / total_count  # Calculating Pew estimate
    pew_estimates[platform_name] = pew_estimate

our_estimate = []

# Print Pew estimates for each social media platform
for platform, estimate in pew_estimates.items():
    print(f"Pew estimate for {platform}: {estimate:.1%}")
    our_estimate.append(f"{estimate:.2f}")

our_estimate = [float(estimate) for estimate in our_estimate]
pew_estimate = [0.29,0.35,0.33,0.47,0.27,0.68,0.30,0.83,0.22,0.22,0.03]

df3['pew_estimate'] = pew_estimate
df3['our_estimate'] = our_estimate
df3['SNS_name']=['WhatsApp','Pinterest','TikTok','Instagram','Snapchat','Facebook','LinkedIn','Youtube',
                 'Twitter','Reddit','BeReal']


import matplotlib.pyplot as plt
from adjustText import adjust_text
import numpy as np

# Scatter plot 생성
plt.scatter(df3['pew_estimate'], df3['our_estimate'])

# 제목과 축 라벨 설정
plt.title('Scatter Plot')
plt.xlabel('PEW_estimate axis')
plt.ylabel('OUR_stimate axis')


# Setting title and labels
plt.title('Scatter Plot')
plt.xlabel('PEW_estimate axis')
plt.ylabel('OUR_estimate axis')
plt.xlim(0,1)
plt.ylim(0,1)
# Initialize a list to hold the text objects

texts = []
for i in range(len(df3)):
    # Add each text annotation to the list
    texts.append(plt.text(df3['pew_estimate'].iloc[i], df3['our_estimate'].iloc[i], df3['SNS_name'].iloc[i], fontsize=9))

# Use adjust_text to automatically adjust the text positions
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='red'))

# prompt: draw the y=x line for scatter plot python

# Add a line for y=x
x_vals = np.linspace(0, 1.2, 100)
y_vals = x_vals
plt.plot(x_vals, y_vals, color='red', linestyle='--')

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is already loaded into df

# Extracting age data
age_counts = df['What is your age (in years)?'].value_counts().sort_index()

# Define age groups with 5-year intervals
age_groups = [(15, 20), (20, 25), (25, 30), (30, 35), (35, 40), (40, 45), (45, 50), (50, 55), (55, 60), (60, 65), (65, 70), (70, 75), (75, 80)]

# Calculate counts for each age group
age_group_counts = {}
for group in age_groups:
    age_group_counts[f'{group[0]}-{group[1]}'] = age_counts.loc[group[0]:group[1]-1].sum()

# Plotting bar graph
plt.figure(figsize=(10, 6))
plt.bar(age_group_counts.keys(), age_group_counts.values())
plt.title('Distribution of Age of Participants')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
